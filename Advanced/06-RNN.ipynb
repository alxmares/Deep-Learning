```python
# No es la aplicación común de RNN, pero este ejemplo demuestra que
# también es funcional apra la clasificación de imágenes.

import os

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train/255.0, x_test/255.0
# Los datos deben ser tratados como sequencia de datos
# 28,28
# input_size=28
# seq_lenght=28 dado que cada secuencia es una fila de la imagen

```

```python
#=== M O D E L ===
model = keras.models.Sequential()
model.add(keras.Input(shape=(28,28))) # seq_lenght,input_size
model.add(layers.SimpleRNN(128, return_sequences=True, activation='relu'))  # por predeterminado tiene la activación tanh
model.add(layers.SimpleRNN(128, return_sequences=False, activation='relu'))
# La shape de estas dos capas no serán la misma. Poner r_s=True hace que la forma
# sea N,28,128 mientras que =False ahce que sea N,128.
# Util para stackers RNNs
model.add(layers.Dense(10))

# No tenemos una activación final, por ello hay que especificar from_logits
print(model.summary())
```

```python
# === L O S S   A N D   O P T I M I Z E R ===
loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
optim = keras.optimizers.Adam(lr=0.001)
metrics = ["accuracy"]

model.compile(loss=loss, optimizer=optim, metrics=metrics)
```

```python
# === T R A I N I N G ===
batch_size=64
epochs=5

model.fit(x_train,y_train, epochs=5, batch_size=batch_size, verbose=2)
```

```python
# === E V A L U A T E ===
model.evaluate(x_test,y_test,batch_size=batch_size,verbose=2)
```
